#**ANALYSIS OF COVID-19 PATIENT DATASET**

Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Reading data from excel file

```python
df = pd.read_excel('Cleaned-Data.xlsx')
```

```python
df.info()
```

```python
df.head()
```

Counting number of persons for different severity of positive tested persons

```python
x = ['Mild', 'Moderate', 'Severe']
y = [df[(df['Severity_Mild']==1)].shape[0], df[(df['Severity_Moderate']==1)].shape[0], df[(df['Severity_Severe']==1)].shape[0]]
```

Visualizing above data using bar plot

```python
fig = plt.figure(figsize=(6,6))
graph = fig.add_subplot()
chart = graph.bar(x, y, width=0.3, color='blue', edgecolor = 'black')

for bar in chart:
    h = bar.get_height()
    graph.text(bar.get_x() + bar.get_width()/2., 1.0*h, '%d' % int(h),ha='center', va='bottom')

plt.minorticks_on()
plt.grid(which='major', linestyle="-", linewidth=0.4)
plt.grid(which='minor', linestyle=":", linewidth=0.3)
plt.title('Results According to Severity', fontsize=20)
plt.xlabel("Severity", fontsize=13)
plt.ylabel("Numbers of People Suffered", fontsize=13)
plt.show()
```

Adding a new column for to find out result of test i.e., 0 for tested negative and 1 for tested positive

```python
df['positive'] = df['Severity_Mild'] + df['Severity_Moderate'] + df['Severity_Severe']
```

```python
df.head(10)
```

Dropping columns that do not play major role in visualization

```python
df = df.drop(axis=1, columns=['Severity_Mild', 'Severity_Moderate', 'Severity_Severe', 'Severity_None', "Contact_Dont-Know", "Contact_No", "Contact_Yes"])
```

```python
df.info()
```

Counting and visualizing total number of positive and negative reported persons

```python
x = ['Positive', 'Negative']
y = [df[(df['positive']==1)].shape[0], df[(df['positive']==0)].shape[0]]
```

```python
fig = plt.figure(figsize=(5,7))
graph = fig.add_subplot()
chart = graph.bar(x, y,  color=['red', 'blue'], width=0.3, edgecolor = 'black')

for bar in chart:
    h = bar.get_height()
    graph.text(bar.get_x() + bar.get_width()/2., 1.0*h, '%d' % int(h),ha='center', va='bottom')

plt.minorticks_on()
plt.grid(which='major', linestyle="-", linewidth=0.4)
plt.grid(which='minor', linestyle=":", linewidth=0.3)

plt.title('Report of Tested Patients', fontsize=20)
plt.xlabel("Result", fontsize=15)
plt.ylabel("Numbers", fontsize=15)
plt.show()
```

Counting and visualizing total number of positive and negative reported persons based on gender

```python
y1 = [df[((df['Gender_Female']==1) & (df['positive']==1))].shape[0], df[((df['Gender_Male']==1) & (df['positive']==1))].shape[0], df[((df['Gender_Transgender']==1) & (df['positive']==1))].shape[0]]
y2 = [df[((df['Gender_Female']==1) & (df['positive']==0))].shape[0], df[((df['Gender_Male']==1) & (df['positive']==0))].shape[0], df[((df['Gender_Transgender']==1) & (df['positive']==0))].shape[0]]
```

```python
df2 = pd.DataFrame(
       [[df[((df['Gender_Female']==1) & (df['positive']==1))].shape[0], df[((df['Gender_Female']==1) & (df['positive']==0))].shape[0]],
        [df[((df['Gender_Male']==1) & (df['positive']==1))].shape[0], df[((df['Gender_Male']==1) & (df['positive']==0))].shape[0]],
        [df[((df['Gender_Transgender']==1) & (df['positive']==1))].shape[0],  df[((df['Gender_Transgender']==1) & (df['positive']==0))].shape[0]]],
       index=['Female', 'Male', 'Transgender'],
       columns = ['Positive', 'Negative']
)
df2
```

```python
df2.plot(kind='barh', figsize=(9,6))

plt.minorticks_on()
plt.grid(which='major', linestyle="-", linewidth=0.4)
plt.grid(which='minor', linestyle=":", linewidth=0.3)

plt.title('Results According to Gender', fontsize=20)
plt.xlabel("Numbers", fontsize=15)
plt.ylabel("Gender", fontsize=15)
plt.show()
```

Finding countries of positive tested persons

```python
df3 = df[(df['positive']==1)]
```

```python
c = set(df3['Country'])
print(c)
```

Counting and visualizing number of positive tested persons for every country

```python
y = []
for item in c:
    y.append(df3[(df3['Country']==item)].shape[0])
```

```python
fig = plt.figure(figsize=(13,7))
graph = fig.add_subplot()
chart = graph.bar(list(c), y, width=0.5, color="blue", edgecolor = 'black')

for bar in chart:
    h = bar.get_height()
    graph.text(bar.get_x() + bar.get_width()/2., 1.0*h, '%d' % int(h),ha='center', va='bottom')

plt.minorticks_on()
plt.grid(which='major', linestyle="-", linewidth=0.4)
plt.grid(which='minor', linestyle=":", linewidth=0.3)
plt.xticks(rotation=90)
plt.title('Results According to Country', fontsize=20)
plt.xlabel("Country", fontsize=15)
plt.ylabel("Numbers", fontsize=15)
plt.show()
```

Counting and visualizing number of positive and negative tested persons by different age groups

```python
df4 = pd.DataFrame(
       [[df[((df['Age_0-9']==1) & (df['positive']==1))].shape[0], df[((df['Age_0-9']==1) & (df['positive']==0))].shape[0]],
        [df[((df['Age_10-19']==1) & (df['positive']==1))].shape[0], df[((df['Age_10-19']==1) & (df['positive']==0))].shape[0]],
        [df[((df['Age_20-24']==1) & (df['positive']==1))].shape[0],  df[((df['Age_20-24']==1) & (df['positive']==0))].shape[0]],
        [df[((df['Age_25-59']==1) & (df['positive']==1))].shape[0],  df[((df['Age_25-59']==1) & (df['positive']==0))].shape[0]],
        [df[((df['Age_60+']==1) & (df['positive']==1))].shape[0],  df[((df['Age_60+']==1) & (df['positive']==0))].shape[0]]],
       index=['Age_0-9', 'Age_10-19', 'Age_20-24', 'Age_25-59', 'Age_60+'],
       columns = ['Positive', 'Negative']
)
df4
```

```python
ax = df4.plot(kind='bar', figsize=(16,8))

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.01, p.get_height() * 1.01))

plt.minorticks_on()
plt.grid(which='major', linestyle="-", linewidth=0.4)
plt.grid(which='minor', linestyle=":", linewidth=0.3)
plt.xticks(rotation=0)
plt.title('Results According to Age-Group', fontsize=20)
plt.xlabel("Age-group", fontsize=15)
plt.ylabel("Numbers", fontsize=15)
plt.show()
```
