##**Refer to company_sales_data.csv and perform the following :**


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv')
df.info()
```

```python
df.head()
```

### a) Read Total profit of all months and show it using a line plot

```python
x = df['month_number']
y = df['total_profit']

plt.plot(x, y, marker = "o" , linewidth=2.0)
plt.xticks(x)
plt.yticks([i for i in range(100000, max(y)+50000, 50000)])
plt.figsize = (10,5)
plt.show()
```

### b) Get Total profit of all months and show line plot with the following Style properties:

*  Line Style dotted and Line-color should be red
*  Show legend at the lower right location.
*  X label name = Month Number
*  Y label name = Sold units number
*  Add a circle marker
*  Line marker color as black
*  Line width should be 3







```python
plt.plot(x, y, 
         label = "Profit of the last year", 
         linestyle=":", 
         color="red", 
         marker="o", markerfacecolor = "black", 
         linewidth=3.0)
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.legend(loc = "lower right")
plt.show()
```

### c) Read all product sales data and show it using a multiline plot

```python
fc = df["facecream"]
fw = df["facewash"]
tp = df["toothpaste"]
bs = df["bathingsoap"]
s = df["shampoo"]
m = df["moisturizer"]

plt.plot(x, fc, label = "facecream", marker="o")
plt.plot(x, fw, label = "facewash", marker="o")
plt.plot(x, tp, label = "toothpaste", marker="o")
plt.plot(x, bs, label = "bathingsoap", marker="o")
plt.plot(x, s, label = "shampoo", marker="o")
plt.plot(x,m, label = "moisturizer", marker="o")

plt.figsize = (30, 15)
plt.legend(loc = "upper left", fontsize=8.5)
plt.xticks(x)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 14000, 16000])
plt.show()
```

### d) Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.scatter(x, tp)
plt.grid(which='major')
plt.show()
```

### e) Read face cream and facewash product sales data and show it using the bar chart

```python
df3 = pd.DataFrame(df[['facecream', 'facewash']], index=df['month_number'])
df3.info()
```

```python
df3.plot(kind = "bar")
plt.figsize = (30,7)
plt.xlabel("Month Number")
plt.ylabel("Number of Products Sold")
plt.show()
```

### f) Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

```python
plt.bar(x, bs, color = "blue")
plt.figsize = (12,8)
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.xticks(x)
plt.yticks([i for i in range(0, max(bs)+2000, 2000)])
plt.show()
plt.savefig('bathingsoap_sales.png', format='png')
```

### g) Read the total profit of each month and show it using the histogram to see most common profit ranges

```python
plt.hist(y, bins=5)
plt.show()
```

### h) Calculate total sale data for last year for each product and show it using a Pie chart

```python
label = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
y = sum(df.iloc(0)).tolist()[1:7]
plt.pie(y, labels=label, autopct = '%1.2f%%', shadow = True)
plt.figsize = (5,5)
plt.show()
```

### i) Read Bathing soap facewash of all months and display it using the Subplot

```python
fig = plt.figure(figsize=(7,5))
fig.subplots_adjust(bottom=0.0, left=0.020, top=1.0, right=0.950)
plt.subplot(2,1,1)
plt.plot(x, bs, label="bathing soap", marker="o", c="red")
plt.legend(loc="lower right")
plt.xticks(x)

plt.subplot(2,1,2)
plt.plot(x, fw, label="facewash", marker="o", c="green")
plt.legend(loc="lower right")
plt.xticks(x)
plt.show()
```
