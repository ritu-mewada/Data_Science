##**Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.**

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('fdata.csv', parse_dates=True, index_col=0)
```

```python
data.plot()
plt.show()
```

## **Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with rendering with a larger grid (major grid) and a smaller grid (minor grid).**

```python
import datetime as dt
import matplotlib.ticker as tck
```

```python
x = list(data.index)
y = data[['Close']]
fig = plt.figure()
graph = fig.add_subplot(111)
graph.plot(x, y , 'r-o')
graph.set_xticks(x)
graph.set_xticklabels([d.strftime("%Y/%m/%d")for d in x])
plt.xlabel('Date')
plt.ylabel('Closing value')
plt.title('Closing stock value of Alphabet Inc.')
graph.yaxis.set_major_locator(tck.MultipleLocator(0.5))
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.6', color ='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color ='black')
plt.show()
```


##**Sample data:**
         Programming languages: Java, Python, PHP, JavaScript, C#, C++
         Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7 

Write a Python programming to display a (vertical) bar chart of the popularity of programming Languages. Use different color for each bar.


```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
y = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos, y , color=['blue', 'yellow', 'green', 'orange', 'cyan', 'red'], edgecolor = 'black')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.xticks(x_pos, x)

plt.minorticks_on()
plt.grid(which='major', linestyle="-", linewidth=0.2, color='black')
plt.grid(which='minor', linestyle=":", linewidth=0.2, color='grey')

plt.show()
```

###Write a Python programming to display a horizontal bar chart of the popularity of programming Languages (for the above data).

```python
plt.barh(x_pos, y, color=['blue', 'yellow', 'green', 'orange', 'cyan', 'red'], edgecolor = 'black')
plt.ylabel('Programming Languages')
plt.xlabel('Popularity')
plt.yticks(x_pos, x)

plt.minorticks_on()
plt.grid(which='major', linestyle="-", linewidth=0.2, color='black')
plt.grid(which='minor', linestyle=":", linewidth=0.2, color='grey')

plt.show()
```

##Write a Python program to create bar plot from a DataFrame.
## Sample Data Frame:
        a b c d e 
        2 4,8,5,7,6
        4 2,3,4,2,6
        6 4,7,4,7,8
        8 2,6,4,8,6
        10 2,4,3,3,2



```python
#Creating a dataframe
df = pd.DataFrame([[4,8,5,7,6], [2,3,4,2,6], 
                   [4,7,4,7,8], [2,6,4,8,6], 
                   [2,4,3,3,2]],
                   index = [2,4,6,8,10],
                   columns = ['a', 'b', 'c', 'd', 'e']
                  )
```

```python
df.plot(kind='bar')

plt.edgecolor = 'black'
plt.minorticks_on()
plt.grid(which='major', linestyle="-", linewidth=0.4, color='green')
plt.grid(which='minor', linestyle=":", linewidth=0.4, color='green')

plt.show()
```

##Write a Python programming to create a pie chart with a title of the popularity of programming Languages.Make multiple wedges of the pie. 
## Sample data:
         Programming languages: Java, Python, PHP, JavaScript, C#, C++
         Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7 

```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
y = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
plt.pie( y, explode=(0.092, 0.0, 0.0, 0.0, 0.0, 0.15),
         labels = x,
         colors = ['blue', 'orange', 'green', 'red', 'violet', 'brown'],
         autopct = '%1.1f%%', 
         shadow = True, 
         startangle = 135,
         wedgeprops = {"edgecolor" : "black", 'linewidth': 1.0}
        )
plt.title("Popularity of Programming Language\n Worldwide," + 
          " Oct 2017 compared to a year ago", 
          bbox={'facecolor':'#CCCCD9', 'edgecolor':'black'})
plt.show()
```
